<h1>NumPyの配列</h1>
NumPyを使うことで、大量のデータを効率よく扱うことができる
数学の計算に有用な多くの機能を持っている
基本は、多数の値が入った配列

```python
import numpy as np


a = [0, 1, 2, 3, 4, 5]
b = np.array(a) # 配列の作成 文字列のみ
print(b) 
```

```python
# linspace関数　範囲を指定して５０で区切る（標準）
x = np.linspace(-5, 5)

print(x)
print('###########')
print(len(x)) # 値の数=50
```

## １次関数

y=2x+1をグラフに描画
$$y=2x+1$$

```python
import matplotlib.pyplot as plt

x = np.linspace(-5, 5) # xを−５〜５の値にする
y = 2*x + 1

plt.plot(x, y)
plt.show()  # グラフを表示する
```

## グラフの装飾
軸のラベル
タイトル
グリッド
凡例

$$y=1.5x$$
$$y=-2x+1$$

```python
x = np.linspace(-3, 3)

y_1 = 1.5*x
y_2 = -2*x + 1

# 軸のラベル
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)

# タイトル
plt.title("linear function")

# グリッド
plt.grid()

# 凡例と線のスタイルを指定
plt.plot(x, y_1, label="y_1")
plt.plot(x, y_2, label="y_2", linestyle="dashed") # linestyle="dashed"点線
plt.legend() # 凡例を表示

plt.show()
```

## 二次関数・三次関数
一次関数、二次関数、三次関数の描画
$$y=2x+1$$
$$y=x^2-4$$
$$y=0.5x^3-6x$$

```python
x = np.linspace(-4, 4)
y_1 = 2*x + 1
y_2 = x**2 - 4
y_3 = 0.5*x**3 - 6*x

plt.plot(x, y_1, label="1st")
plt.plot(x, y_2, label="2nd")
plt.plot(x, y_3, label="3rd")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```
