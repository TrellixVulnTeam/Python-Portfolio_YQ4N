```python
# GroupByによる集計
import pandas as pd


df = pd.read_excel('group_by.xlsx')
display(df)
```

```python
df.groupby('担当者').mean()
```

```python
df.groupby('曜日').max(numeric_only=True)
```

```python
# インデックス変更
df.index = range(1, 10)
display(df)
```

```python
# カラム名変更
df.columns = ['data', 'week', 'parson', 'sales']
display(df)
```

```python
df_new = df.set_index('parson')
display(df_new) #引数に inplace=True とすると df 自体も変化してしまう
```

```python
# インデックスのリセット
df_new = df.reset_index()
display(df_new) 
```
