```python
from pandas_datareader import data
# データ改正を支援する機能を提供 pandas 1列Series 2列DataFrame
import pandas as pd
# グラフを書くためのライブラリ
import matplotlib.pyplot as plt
# JupyterLab上で作成したグラフを表示
%matplotlib inline
import numpy as np
```

```python
start = '2021-01-01'
end = '2021-12-31'
# 日経平均のデータをyahooから期間を指定して取得
df = data.DataReader('^N225', 'yahoo', start, end)
```

```python
# 上位１０件の確認
df.head(10)
```

```python
# インデックスの確認
df.index
```

```python
data = df.index
price = df['Adj Close']
```

```python
# マットプロットリブを使ってグラフ化
plt.plot(data, price)
```

```python
# グラフのサイズ変更
plt.figure(figsize=(30, 10)) #(横, 縦)
plt.plot(data, price, label='Nikkei225')
# タイトル表示
plt.title('N225', color='blue', backgroundcolor='white', size=40, loc='center')
plt.xlabel('data', color='black', size=30)
plt.ylabel('price', color='black', size=30)
# 右上に凡例を表示
plt.legend()
```

```python
# 移動平均線の作成
# ３つの期間設定

span1 = 5
span2 = 25
span3 = 75

df['sma1'] = price.rolling(window = span1).mean()
df['sma2'] = price.rolling(window = span２).mean()
df['sma3'] = price.rolling(window = span3).mean()
# sma1=カラム名 window=期間　meam()=平均算出
```

```python
pd.set_option('display.max_rows', None) #表示が省略されるのを防ぐためのもの
df.head(100)
```

```python
# グラフ化する
plt.figure(figsize=(30, 10))
plt.plot(data, price, label='Nikkei225')
plt.plot(data, df['sma1'], label='short')
plt.plot(data, df['sma2'], label='middle')
plt.plot(data, df['sma3'], label='long')
plt.title('N225', color='blue', backgroundcolor='white', size=40, loc='center')
plt.xlabel('data', color='black', size=30)
plt.ylabel('price', color='black', size=30)
plt.legend()
```

```python
# まとめ
from pandas_datareader import data

start = '2021-01-01'
end = '2021-12-31'
company_code = '9036.jp'

df = data.DataReader(company_code, 'stooq', start, end)

data = df.index
price = df['Close']

span1 = 5
span2 = 25
span3 = 75

df['sma1'] = price.rolling(window = span1).mean()
df['sma2'] = price.rolling(window = span２).mean()
df['sma3'] = price.rolling(window = span3).mean()

plt.figure(figsize=(30, 15))

plt.subplot(2, 1, 1)
plt.plot(data, price, label='Keeper', color='#C6D57E')
plt.plot(data, df['sma1'], label='short',color='#D57E7E')
plt.plot(data, df['sma2'], label='middle', color='#A2CDCD')
plt.plot(data, df['sma3'], label='long', color='#FFE1AF')
plt.legend()

plt.subplot(2, 1, 2)
plt.bar(data,df['Volume'], label='Volume', color='grey')
plt.legend()
```

```python
# 関数化
def graphing_stock(company_code, start, end):
    from pandas_datareader import data　# 内部に記述しないとエラーになったためこちらに記載
    df = data.DataReader(company_code, 'stooq')
    df = df[(df.index>=start) & (df.index<=end)]
    
    data = df.index
    price = df['Close']
     
    span1 = 5
    span2 = 25
    span3 = 75

    df['sma1'] = price.rolling(window = span1).mean()
    df['sma2'] = price.rolling(window = span２).mean()
    df['sma3'] = price.rolling(window = span3).mean()

    plt.figure(figsize=(30, 15))

    plt.subplot(2, 1, 1)
    plt.plot(data, price, label='company_code', color='#C6D57E')
    plt.plot(data, df['sma1'], label='short',color='#D57E7E')
    plt.plot(data, df['sma2'], label='middle', color='#A2CDCD')
    plt.plot(data, df['sma3'], label='long', color='#FFE1AF')
    plt.legend()

    plt.subplot(2, 1, 2)
    plt.bar(data,df['Volume'], label='Volume', color='grey')
    plt.legend()
```

```python
graphing_stock('4661.jp', '2021-01-01', '2021-12-31')
```
